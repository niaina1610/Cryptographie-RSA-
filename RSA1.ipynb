```python
from sympy import isprime

def recherche_p_et_q(N):
    i = 1
    while True:
        p = 2**i - 1  # Nombre de Mersenne
        if N % p == 0 and isprime(p):  # Vérifie si p est premier
            q = N // p  # Division entière
            return p, q
        i += 1

# Entrée de l'utilisateur
N = int(input("Entrer le nombre N : "))
p, q = recherche_p_et_q(N)
print(f"p = {p}, q = {q}")
```

```python
#algorithme d'euclide etendu
#ceci est utile si on veut trouver d pour le decryptage
def modinv(ua,n):
    t,newt = 0,1
    r,newr = n,ua
    while newr !=0:
        q = r // newr
        t,newt = newt, t - q * newt
        r,newr = newr, r - q * newr
    if r > 1:
        raise ValueError(f"{ua} n'est pas inversible modulo {n}")
    if t < 0:
        t+= n
    return t
```

```python
#decryptage et encryptage
n = int(input("Entrer le nombre n : "))
k = int(input("Entrer l'exposant : "))#clé privé pour dechiffrer et clé public pour encrypter
# d obtenu par inverse modulaire de e avec F si c'est pour decrypter
M = int(input("Entrer le nombre à décrypter ou crypter : "))
N = pow(M, k, n)
print("Le résultat est :", N)
```
